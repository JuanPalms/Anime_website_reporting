### The top:

- Best scored anime:

```python
datos.loc[datos['score'].idxmax()]
```

- Top 3 longest and popular animes (In the same emission period)

```python
top3_longest_anime=datos\
    .sort_values(by="number_episodes", ascending=False)\
    .head(15)\
    .sort_values(by="members",ascending=False)\
    .head(3)
top3_longest_anime
```

- Top ten Most popular anime

```python
most_popular_anime=datos\
    .sort_values(by="members", ascending=False)\
    .head(10)
most_popular_anime
```

-  Top ranked studios

```python
top_ranked_studios=\
    datos[['studio','score']]\
    .groupby('studio')\
    .mean()\
    .sort_values(by="score",ascending=False)\
    .head(10)
top_ranked_studios
```

- Most popular studio

```python
top_10_popular_studios=\
    datos[['studio','members']]\
    .groupby('studio')\
    .mean()\
    .sort_values(by="members", ascending=False)\
    .head(10)
top_studios_list = top_10_popular_studios.index.tolist()
animes_from_top_studios = datos[datos['studio'].isin(top_studios_list)]
```

```python
top_10_popular_studios
```

- Studio with the greatest number of episodes

```python
top_10_more_episodes_studios=\
    datos[['studio','number_episodes']]\
    .groupby('studio')\
    .sum()\
    .sort_values(by="number_episodes", ascending=False)\
    .head(10)
#top_studios_list = top_10_popular_studios.index.tolist()
#animes_from_top_studios = datos[datos['studio'].isin(top_studios_list)]
```

- Top 10 popular demographics

```python
top10_popular_demografics=datos[['demographics','members']]\
    .groupby("demographics")\
    .sum()\
    .sort_values(by="members", ascending = False)
top10_popular_demografics
```
